# K Means Clustering Visualization

In this notebook, we will create a simple visualization of K Means clustering to deepen our understanding of how it works.

We will create some points that fall in 2 distinct clusters and visualize how they are grouped.

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import numpy as np
from sklearn.cluster import KMeans

X = np.array([[1, 2],
              [1.5, 1.8],
              [5, 8],
              [8, 8],
              [1, 0.6],
              [9, 11]])

plt.scatter(X[:,0], X[:,1], s=100, linewidths=5, c='b')
plt.show()
```

Now we will fit the K Means model and visualize it. The Xs on the graph wil represent the centroids used to determing the clusters, which will be visualized as different colors.

```python
clf = KMeans(n_clusters=2)
clf.fit(X)

centroids = clf.cluster_centers_
labels = clf.labels_

colors = ['r.', 'b.', 'g.', 'c.', 'k.', 'm.']

for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize=25)

plt.scatter(centroids[:,0], centroids[:,1], marker ='x', s=100, linewidth=5, c='g')
plt.show()
```
